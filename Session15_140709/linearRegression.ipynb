# Regression and Model Fitting

### A different approch

We would like to create a model which describes our data set. In previous discussions, we tried to understand the underlying PDF of our data. We will step back and take a different approch now. We will take our data, and paint labels on it. Data will now have independent variables, often denoted as the set x, which describe corresponding dependent variables, denoted as y's. 

The model we construct of this data set will be a function which will describe the expectation value of y given x. We can use this model to make predictions for other values of y given x's we have not yet measured, or to better understand parameters. This will be much easier to understand given a data set to look at.

Here is a made up data set of the distance modulus of fake supernova at different red shifts. The data have heteroscedastic errors drawn from a guassian.

### An example: Supernovae, their Distance Modulus and Red shift

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import lognorm

from astroML.cosmology import Cosmology
from astroML.datasets import generate_mu_z
from astroML.linear_model import LinearRegression, PolynomialRegression,\
    BasisFunctionRegression
from astroML.plotting import setup_text_plots


#------------------------------------------------------------
# Generate data
z_sample, mu_sample, dmu = generate_mu_z(100, random_state=0)

cosmo = Cosmology()
z = np.linspace(0.01, 2, 1000)
mu_true = np.asarray(map(cosmo.mu, z))
```

```python
ax=plt.subplot(111)
ax.errorbar(z_sample, mu_sample, dmu, fmt='.k', ecolor='gray', lw=1)
ax.set_ylabel("distance modulus")
ax.set_xlabel("z")
plt.show()
```

To start using some of the new jargon, the each point has a y (distance modulus) and x (red shift).

Now we would like to make a model which describes this dataset. It will be a function that returns a prediction for distance modulus given red shift and some parameters of this model. There's a zoo of different models you can try to create. Lets keep it simple to start off with, and make a linear model. They have the following form

$$f(x|\theta) = \sum^{k}_{p=1} \theta_{p}g_{p}(x)$$

to make our model, we take a function of each of the k independent variables (the g(x)'s), weight them by thier own constants (the thetas) and sum them up.

Interesting things to note: the g(x)'s do not need to be linear functions. you can use $$f(x,\theta)= \theta*sin(x)$$

That is a perfectly linear model. but the g(x)'s cannot depend on any of the theta's. That is g(x) is g(x), and not g(x,theta). $$f(x,\theta)= \theta*sin(\theta*x)$$ is nonlinear model

To find the model that best describes our data, or the best fit model, our task is then to choose the right values for the theta's

### Finding The Best Parameters

Let's consider the simplest case. Let's try to fit a straight line to our data.$$f(x|\theta) = \theta_{0} + \theta_{1} x$$Now how do we choose the best values of theta. Lets assume that the errors in x are negligable and assume the y values deviation from the model follows from a gausian. Then$$p(y_{i}|x_{i}, \theta, I) \propto exp({-[y_{i}-f(x_{i}|\theta)^2 \over 2\sigma_{i}^2})$$ skipping ahead to the log(L) $$Ln(L) \propto \sum^{N}_{i=1}({-(y_{i} - (\theta_{0} + \theta_{1}x))^2 \over 2\sigma_{i}^{2}})$$This is maximized when y is close to the prediction of the model. So we effectively want to reduce the squared error.

It turns out there is a vectorized approch to quickly solve for all of the theta's at once. So lets take a break to talk about vectorizing the above discussion.

### A Vectorized Approch

Lets repackage our data. If we have N data points, we can group the y's, x's, and theta's in their own matricies as well as C, a covariance matrix for the errors$$ \mathbf{Y}=\begin{bmatrix}
y_{0}\\
y_{1}\\
y_{2}\\
.\\
y_{N-1}
\end{bmatrix}
\mathbf{\Theta}=\begin{bmatrix}
\theta_{0}\\
\theta_{1}\\
\end{bmatrix}
\mathbf{M}=\begin{bmatrix}
1 && x_{0}\\
1 && x_{1}\\
1 && x_{2}\\
. && .\\
1 && x_{N-1}\\
\end{bmatrix}
\mathbf{C}=\begin{bmatrix}
\sigma_{0}^{2} && 0 && . && 0\\
0 && \sigma_{1}^2 && . && 0\\
. && . && . && .\\
0 && 0 && . && \sigma_{N-1}^{2}
\end{bmatrix}$$

The benefit is that now we can neatly express the function we are trying to reduce $$(\mathbf{Y}-\mathbf{M}\mathbf{\Theta})^{T}(\mathbf{Y}-\mathbf{M}\mathbf{\Theta})$$

The Maximum Liklihood solution for this regression is given by the "normal equation"

$$\mathbf{\Theta} = (M^{T}C^{-1}M)^{-1}(M^{T}C^{-1}Y)$$

wow, cool! Let's do this by hand on our super nova data

```python
Y=np.vstack(mu_sample)
ones=np.ones(len(z_sample))
M=np.hstack((np.vstack(ones),np.vstack(z_sample)))
C=np.zeros((100,100))
np.fill_diagonal(C,dmu)
```

```python
Cinv=np.linalg.pinv(C)
left=np.dot(M.T,np.dot(Cinv,Y))
right=np.linalg.pinv(np.dot(M.T,np.dot(Cinv,M)))
Theta=np.dot(right,left)
Theta
```

How did we do? Lets use our model and feed in a range of z values as inputs and weight them by the theta's we found to make a prediction.

```python
m=np.hstack((np.ones(len(z))[:,None],z[:,None]))
y_predic=np.dot(m,Theta)
ax=plt.subplot(111)
ax.errorbar(z_sample, mu_sample, dmu, fmt='.k', ecolor='gray', lw=1)
ax.plot(z,y_predic)
ax.set_ylabel("distance modulus")
ax.set_xlabel("z")
plt.show()
```

Although we can express Theta in just one line of math, it takes a few lines more to compute it in python-ese. the astroML package is able to make a fit and predictions pretty painlessly

```python
from astroML.linear_model import LinearRegression
model = LinearRegression()
model.fit(z_sample[:,None],mu_sample,dmu)
y_pred2 = model.predict(z[:,None])
ax=plt.subplot(111)
ax.errorbar(z_sample, mu_sample, dmu, fmt='.k', ecolor='gray', lw=1)
ax.plot(z,y_pred2)
ax.set_ylabel("distance modulus")
ax.set_xlabel("z")
plt.show()
model.coef_
```

### Polynomial Regression

The straight line was an okay fit, but it looks like we can do better. Lets try polonomial fits instead. The recipie is almost exactly the same, we just need to update our matrix M and our matrix Theta. For a second degree polynomial we would have 
$$\mathbf{\Theta}=\begin{bmatrix}
\theta_{0}\\
\theta_{1}\\
\theta_{2}\\
\end{bmatrix}
\mathbf{M}=\begin{bmatrix}
1 && x_{0} && x_{0}^2\\
1 && x_{1} && x_{1}^2\\
1 && x_{2} && x_{2}\\
. && .\\
1 && x_{N-1} && x_{N-1}^2\\
\end{bmatrix}$$
AstroML can also solve polynomial regression, in the same spirit as linear regression

```python
from astroML.linear_model import PolynomialRegression
modelPoly=PolynomialRegression(4)
modelPoly.fit(z_sample[:,None],mu_sample,dmu)
y_predPoly= modelPoly.predict(z[:,None])

ax=plt.subplot(111)
ax.errorbar(z_sample, mu_sample, dmu, fmt='.k', ecolor='gray', lw=1)
ax.plot(z,y_predPoly)
ax.set_ylabel("distance modulus")
ax.set_xlabel("z")
plt.show()
```

### Nonlinear Functions of X

So far our models have consisted of linear functions of x. we can still do linear regression with functions that are nonlinear in x, provided that they are still linear in the thetas. For example, our model can be a basis of gausians, trigonometric functions, inverse quadratic functions, and splines. Lets try fitting a basis of gausians. to do this, we will need to provide the appropriate astroML package with some parameters, namely the number of gausians and their the mean and width.

```python
from astroML.linear_model import BasisFunctionRegression

basis_mu = np.linspace(0, 2, 15)[:, None] #linearly space 15 gausians, with centers between 0 and 2
basis_sigma = 3 * (basis_mu[1] - basis_mu[0]) #the 'width' of the gausian is three times the spacing between them
modelBasis = BasisFunctionRegression('gaussian',mu=basis_mu, sigma=basis_sigma)
modelBasis.fit(z_sample[:,None],mu_sample,dmu)
y_predBasis = modelBasis.predict(z[:,None])

ax=plt.subplot(111)
ax.errorbar(z_sample, mu_sample, dmu, fmt='.k', ecolor='gray', lw=1)
ax.plot(z,y_predBasis)
ax.set_ylabel("distance modulus")
ax.set_xlabel("z")
ax.set_xlim(0.01, 1.8)
ax.set_ylim(36.01, 48)
plt.show()
```

### Overfitting and Peanalizing the Regression

As we've added degrees of freedom to our models we were able to achieve better and better fits. Of course, if we got carried away and added many degrees of freedom we could keep making the fit better still. The trouble with this is that the resulting model may not be physically meanigful. For example, in our last two models for the distance modulus, our model drops off at the end dramatically. This is probably not corrrect. 

To correct for this, we preform "Regularization" by peanalizing the theta's by a constant. This has the effect of disallowing large thetas.

The text discusses two methods that use regularization in detail. Lets first discuss ridge regression

Instead of minimizing the least squares, we are now minimizing 

$$(\mathbf{Y}-\mathbf{M}\mathbf{\Theta})^{T}(\mathbf{Y}-\mathbf{M}\mathbf{\Theta}) + \lambda(\mathbf{\Theta}^{T}\mathbf{\Theta}) $$ 

The normal equation must also be adjusted

$$\mathbf{\Theta} = (M^{T}C^{-1}M + \lambda I)^{-1}(M^{T}C^{-1}Y)$$

where I is the identity matrix

```python
from sklearn.linear_model import Ridge

# Manually convert data to a gaussian basis
#  note that we're ignoring errors here, for the sake of example.
def gaussian_basis(x, mu, sigma):
    return np.exp(-0.5 * ((x - mu) / sigma) ** 2)

centers = np.linspace(0, 1.7, 15) #as before, there are 15 gausians, centered evenly between 0 and 2.0
widths = basis_sigma #use the same width as before
X = gaussian_basis(z_sample[:, np.newaxis], centers, widths) #wrap all the gausians in a maxtrix

modelRidge = Ridge(alpha = 0.0) #confusingly, by alpha they mean lambda
modelRidge.fit(X, mu_sample)
y_predRidge = modelRidge.predict(gaussian_basis(z[:, None], centers, widths))

w = modelRidge.coef_
curves = 37 + w * gaussian_basis(z[:, np.newaxis], centers, widths)
curves = curves[:, abs(w) > 0.01]

ax=plt.subplot(111)
#ax.plot(z, curves, c='gray', lw=1, alpha=0.5)
ax.plot(z, y_predRidge)
ax.errorbar(z_sample, mu_sample, dmu, fmt='.k', ecolor='gray', lw=1, ms=4)
ax.set_xlim(0.01, 2.0)
ax.set_ylim(36, 52)
plt.show()
```

Notice how we supressed the over fitting here as opposed to the previous example. the fit looks particularly better at the end of the data.

Another approch is to use LASSO (Least absolute shrinkgage and Selection). Now we penalize the absolute value of the theta's, instead of their square.

$$(\mathbf{Y}-\mathbf{M}\mathbf{\Theta})^{T}(\mathbf{Y}-\mathbf{M}\mathbf{\Theta}) + \lambda | \mathbf{\Theta} | $$

LASSO will actually zero out some theta's. The effect is to make a more "sparse model" and to reduce the number of "knobs to turn" in your model. To emphisize this point, we will use a basis of 100 gausians, do ridge regression, and then use a basis of 100 gausians using LASSO instead, and examine the coefficents at the end.

```python
np.random.seed(0)

z_sample, mu_sample, dmu = generate_mu_z(100, random_state=0)
cosmo = Cosmology()

z = np.linspace(0.01, 2, 1000)
mu = np.asarray(map(cosmo.mu, z))

from sklearn.linear_model import LinearRegression, Ridge, Lasso
centers = np.linspace(0, 1.8, 100)
widths = 0.2
X = gaussian_basis(z_sample[:, np.newaxis], centers, widths)

#------------------------------------------------------------
# Set up the figure to plot the results
fig = plt.figure(figsize=(15, 10))
fig.subplots_adjust(left=0.1, right=0.95,
                    bottom=0.1, top=0.95,
                    hspace=0.15, wspace=0.2)

classifier = [LinearRegression, Ridge, Lasso]
kwargs = [dict(), dict(alpha=0.005), dict(alpha=0.001)]
labels = ['Linear Regression', 'Ridge Regression', 'Lasso Regression']

for i in (1,2):
    clf = classifier[i](fit_intercept=True, **kwargs[i])
    clf.fit(X, mu_sample)
    w = clf.coef_
    fit = clf.predict(gaussian_basis(z[:, None], centers, widths))

    # plot fit
    ax = fig.add_subplot(231 + i)
    ax.xaxis.set_major_formatter(plt.NullFormatter())

    # plot curves for regularized fits
    if i == 0:
        ax.set_ylabel('$\mu$')
    else:
        ax.yaxis.set_major_formatter(plt.NullFormatter())
        curves = 37 + w * gaussian_basis(z[:, np.newaxis], centers, widths)
        curves = curves[:, abs(w) > 0.01]
        ax.plot(z, curves,
                c='gray', lw=1, alpha=0.5)

    ax.plot(z, fit, '-k')
    ax.plot(z, mu, '--', c='gray')
    ax.errorbar(z_sample, mu_sample, dmu, fmt='.k', ecolor='gray', lw=1, ms=4)
    ax.set_xlim(0.001, 1.8)
    ax.set_ylim(36, 52)
    ax.text(0.05, 0.93, labels[i],
            ha='left', va='top',
            bbox=dict(boxstyle='round', ec='k', fc='w'),
            transform=ax.transAxes)

    # plot weights
    ax = plt.subplot(234 + i)
    ax.xaxis.set_major_locator(plt.MultipleLocator(0.5))
    ax.set_xlabel('$z$')
    if i == 0:
        ax.set_ylabel(r'$\theta$')
        w *= 1E-12
        ax.text(0, 1.01, r'$\rm \times 10^{12}$',
                transform=ax.transAxes)
    ax.scatter(centers, w, s=9, lw=0, c='k')

    ax.set_xlim(-0.05, 1.8)

    if i == 1:
        ax.set_ylim(-2, 4)
    elif i == 2:
        ax.set_ylim(-0.5, 2)

    ax.text(0.05, 0.93, labels[i],
            ha='left', va='top',
            bbox=dict(boxstyle='round', ec='k', fc='w'),
            transform=ax.transAxes)

plt.show()
```
